```python
import pandas as pd

df = pd.DataFrame({' Country ' : ['India', 'Indonesia', 'Malaysia', 'Indonesia', 'Jepang' ],
                  'Age ' : [44 , 23, 45, 20, 21],
                  ' salary' : [72000, 45000, 56000, 66000, 42000]})

df
```

```python
#Label encoding pada scikit learn

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df[' Country '] = label_encoder.fit_transform(df[' Country '])
df
```

```python
label_encoder.classes_ #ini sebab kenapa bisa mengahsilkan angka 0,1,2 ... ,di urutkan berdasarkan alfabet. kelemahan jika ada sebuah pengkondisian
```

```python
#ONE HOT ENCODING

df = pd.DataFrame({'Country' : ['India', 'Indonesia', 'Malaysia', 'Indonesia', 'Jepang' ],
                  'Age' : [44 , 23, 45, 20, 21],
                  'salary' : [72000, 45000, 56000, 66000, 42000]})

df
```

```python
x = df[' Country '].values.reshape(-1,1)
x
```

```python
from sklearn.preprocessing import OneHotEncoder

onehot_encoder = OneHotEncoder()
X = onehot_encoder.fit_transform(x).toarray()
X
```

```python
onehot_encoder.categories_ #karena ada 3 buah isi list yang berbeda maka ada 3 digit bits
```

```python
df_onehot = pd.DataFrame(X, columns=[str(i) for i in range(X.shape[1])])
df_onehot
```

```python
df = pd.concat([df_onehot, df], axis=1)
df
```
