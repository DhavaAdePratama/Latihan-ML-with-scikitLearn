```python
corpus = [
    "This is the first document.",
    "This document is the second document.",
    "And this is the third one.",
    "Is this the first document?",
]
corpus
```

```python
# TF - IDF Weights dengan tfidVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words= 'english')
X = vectorizer.fit_transform(corpus)

print(X)
# bagian kanan adalah nilai score semakin besar semakin bagus, sedangkan sebelah kiri adalah index pada baris serta kata apa yg terdapat pada kalimat tersebut
```

```python
print(vectorizer.get_feature_names_out()) #melihat kata apa saja yang menjadi point
```

```python
# Mempresentasikan bobot nilai nya

X.todense()
```

```python
import pandas as pd
df = pd.DataFrame(X.todense().T, # T = transpose
                  index=vectorizer.get_feature_names_out(),
                  columns=[f'D{i+1}' for i in range(len(corpus))])
df #nilai hasil merupakan nilai yang telah di normalisasi menggunakan l2_normalisation
```
